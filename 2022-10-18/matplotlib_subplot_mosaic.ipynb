```python
import xarray as xr

from matplotlib import pyplot as plt
```

```python
da = xr.DataArray([1, 2, 3, 1])
```

```python
da.plot()
```

```python
plt.figure()

plt.subplot(1, 2, 1)

plt.plot(da)

plt.subplot(1, 2, 2)

plt.plot(-da)
```

```python
plt.figure()

plt.subplot(1, 2, 1)

da.plot()

plt.subplot(1, 2, 2)

(-da).plot()
```

```python
fig, ax = plt.subplots(1, 2)

ax[0].plot(da)

ax[1].plot(-da)
```

```python
fig, ax = plt.subplots(1, 2)

da.plot(ax=ax[0])

(-da).plot(ax=ax[1])
```

```python
# fig, axs = plt.subplot_mosaic([['a', 'a', 'b'], ['a', 'a', '.']], constrained_layout=True)

fig, axs = plt.subplot_mosaic(
    """
    aabb
    aabb
    """,
    layout="constrained",
)

da.plot(ax=axs['a'])

(-da).plot(ax=axs['b'])
```

Let's read some data from: https://doi.pangaea.de/10.1594/PANGAEA.890943

```python
import pangaeapy
```

```python
dataset = pangaeapy.pandataset.PanDataSet("https://doi.pangaea.de/10.1594/PANGAEA.890943")
print(dataset.title)
```

```python
dataset.data.head(5)
```

```python
data = dataset.data
data
```

```python
closest_to_surface = data.groupby("Event").first()
closest_to_surface
```

```python
closest_to_surface.plot.scatter(x="Longitude", y="Latitude", c="Temp", s=15, cmap="viridis", figsize=(12, 7));
```

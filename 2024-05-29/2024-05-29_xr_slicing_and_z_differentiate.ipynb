```python
import xarray as xr
import numpy as np
```

```python
data = xr.DataArray(
    np.random.normal(size=(12, 100, 180, 360)),
    dims=("t", "z", "y", "x"),
)
data = data.assign_coords(z=np.linspace(2, 998, 100))
data
```

```python
data.interp(z=np.linspace(0, 500, 25), method="linear").isel(t=0, x=0, y=0)
```

```python
data.interp(z=np.linspace(0, -500, 25), method="linear").isel(t=0, x=0, y=0)
```

```python
(
    data
    .interp(z=np.linspace(0, 500, 25), method="linear")
    .isel(t=0, x=0, y=0)
    .differentiate("z")
)
```

```python
data.interp(z=xr.DataArray(np.linspace(0, 500, 25), dims=("depth", )), method="linear")
```

```python
(
    data
    .interp(z=xr.DataArray(np.linspace(0, 500, 25), dims=("depth", )), method="linear")
    .isel(t=0, x=0, y=0)
    .differentiate("depth")
)
```

```python
data.isel(z=slice(0, 1)) + data.isel(z=slice(7, 8))
```

```python
data.sel(z=slice(333.4, 700)).diff("z") / data.sel(z=slice(333.4, 700)).z.diff("z")
```

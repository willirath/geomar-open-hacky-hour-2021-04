```python
import xarray as xr
```

```python
airtemps = xr.tutorial.load_dataset('air_temperature')
```

```python
temp = airtemps.air
temp
```

```python
temp.var("time").plot()
```

```python
temp_reg = temp.where(
    (temp.coords["lon"] > 300)
    & (temp.coords["lon"] < 310)
    & (temp.coords["lat"] > 50)
    & (temp.coords["lat"] < 60)
)
```

```python
temp_reg
```

```python
temp_reg.var("time").plot()
```

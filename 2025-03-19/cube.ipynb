```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot([0, 1], [0, 1], [0, 1], 'r')

for nargor in range(70):
    x, y = list(np.random.uniform(0, 1, size=(2, )))
    ax.plot(
        [x, x], [y, y],
        [1, 0], 'm',
        alpha=0.1
    )
    ax.plot(
        [x, ], [y, ],
        [1, ], 'mo',
        zorder=1000
    )

x = np.linspace(0, 1, 11)
y = np.linspace(0, 1, 21)
z = np.array([0, ])
xx = x[:, np.newaxis] + 0 * y[np.newaxis, :]
yy = 0 * x[:, np.newaxis] + y[np.newaxis, :]
zz = np.zeros_like(xx) + 1

# colors = plt.get_cmap(cmap)(norm(array))

ax.plot_surface(
    xx, yy, zz,
    facecolors=plt.get_cmap("viridis")(Normalize()(np.cos(10 * xx) * np.sin(5 * yy)))
)

ax.plot(*p.boundary.xy, [1, ], 'orange', zorder=100)

ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.1, 1.1)
ax.set_zlim(-0.1, 1.1)
```

```python
import numpy as np
import xarray as xr
```

```python
data = xr.DataArray(
    np.random.normal(size=(100, 200)), dims=("x", "y"),
)
for _nsmooth in range(40):
    data = data.rolling(x=3, center=True, min_periods=1).mean()
    data = data.rolling(y=3, center=True, min_periods=1).mean()
data.plot()
```

```python
data = data.where(data.x > 50)
data.plot()
```

```python
data_masked = (data > 0.1).astype(int)
data_masked = data_masked.where(data_masked.x > 50)

data_masked.plot()
data_masked.plot.contour(colors="cyan", levels=[0.5, ], linewidth=2)
```

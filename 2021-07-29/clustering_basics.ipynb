```python
import sklearn
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
Ncluster = 3
Npercluster = 100
csep = 100
cwidth = 10

xc = csep * np.random.normal(size=(Ncluster, ))
yc = csep * np.random.normal(size=(Ncluster, ))

x = (cwidth * np.random.normal(size=(Npercluster, Ncluster)) + xc).flatten()
y = (cwidth * np.random.normal(size=(Npercluster, Ncluster)) + yc).flatten()
```

```python
plt.plot(xc, yc, 'ko')
plt.plot(x, y, 'k.')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=Ncluster)
```

```python
X = np.stack((x, y)).T
print(X.shape)
```

```python
km.fit(X)
```

```python
cluster_id = km.predict(X)
```

```python
plt.scatter(x, y, c=cluster_id)
```

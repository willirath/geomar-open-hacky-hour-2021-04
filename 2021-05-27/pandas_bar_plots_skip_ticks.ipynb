```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(
    np.random.randint(0, 100, size=(100, )),
    index=np.arange(100),
)

display(s)
```

```python
ax = s.plot.bar(figsize=(18, 4))

ax.xaxis.set_ticks(s.index[::10]);
ax.xaxis.set_ticklabels(map(str, s.index[::10]));
```

```python
dates = pd.date_range("2021-01-01", "2021-12-31", freq="D")[:100]
dates
```

```python
s = pd.Series(
    np.random.randint(0, 100, size=(len(dates), )),
    index=dates
)

display(s)
```

```python
ax = s.plot.bar(figsize=(18, 4))

ax.xaxis.set_ticks(np.arange(100)[::10]);
ax.xaxis.set_ticklabels(dates[::10]);
```

See https://stackoverflow.com/a/19387765 for extracting all info from the ticks themselves.

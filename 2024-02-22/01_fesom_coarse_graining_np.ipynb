```python
import xarray as xr
import numpy as np
```

```python
# parameters
num_nodes = 11_000
num_neighbors = 10
len_time = 400
len_z = 40
```

```python
n = xr.DataArray(
    np.arange(num_nodes, ),
    dims=("n", ),
    name="n",
)
n
```

```python
neighbors = xr.DataArray(
    np.arange(num_neighbors),
    dims=("j", ),
    name="neighbor",
)
```

```python
z = xr.DataArray(
    np.arange(len_z),
    dims=("z", ),
    name="z",
)
z
```

```python
groups = (
    n.isel(n=slice(None, None, num_neighbors))
    + neighbors
)
groups
```

```python
T = xr.DataArray(
    np.random.normal(size=(num_nodes, len_time, len_z)),
    dims=("n", "t", "z"),
    name="temp",
    coords={"n": n, "z": z},
)
T
```

```python
W = xr.ones_like(n * neighbors)
W = W.assign_coords(n=n)
W
```

```python
(T * W).sel(n=groups).sum("j") / W.sel(n=groups).sum("j")
```

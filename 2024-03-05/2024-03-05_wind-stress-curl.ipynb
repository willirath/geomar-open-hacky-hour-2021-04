```python
winds = xr.open_mfdataset(
    "/home/jovyan/shared_data/incoming_nesh/WIND_GLO_PHY_CLIMATE_L4_MY_012_003/cmems_obs-wind_glo_phy_my_l4_P1M_202211/2000/cmems_obs-wind_glo_phy_my_l4_P1M_2000??.nc"
)
# winds = winds.mean("time")
winds
```

```python
winds = winds.isel(
    lon=slice(None, None, 1),
    lat=slice(None, None, 1),
)
winds
```

```python
# winds.squeeze().plot.quiver(
#     x="lon", y="lat",
#     u="eastward_stress",
#     v="northward_stress",
# )
```

```python
taux = winds.eastward_stress
tauy = winds.northward_stress
```

```python
winds.lat
```

```python
A = 6371.0e3
A
```

```python
delta_y = np.deg2rad(winds.lat.shift(lat=-1) - winds.lat) * A
delta_y
```

```python
delta_x = (
    np.deg2rad(winds.lon.shift(lon=-1) - winds.lon) * A
    * np.cos(np.deg2rad(winds.lat))
)
delta_x
```

```python
(delta_x / delta_y).T.plot()
```

```python
dtauy_by_dx = (tauy.shift(lon=-1) - tauy) / delta_x
dtauy_by_dx = (dtauy_by_dx.shift(lat=-1) + dtauy_by_dx) / 2.0
dtauy_by_dx
```

```python
dtaux_by_dy = (taux.shift(lat=-1) - taux) / delta_y
dtaux_by_dy = (dtaux_by_dy.shift(lon=-1) + dtaux_by_dy) / 2.0
dtaux_by_dy
```

```python
curlz_tau = (dtauy_by_dx - dtaux_by_dy)
curlz_tau = curlz_tau.assign_coords(
    lat=(winds.lat.shift(lat=-1) + winds.lat) / 2,
    lon=(winds.lon.shift(lon=-1) + winds.lon) / 2,
)
curlz_tau
```

```python
curlz_tau.plot.hist(bins=31);
```

```python
curlz_tau = curlz_tau.isel(lon=slice(0, -1), lat=slice(0, -1))
curlz_tau
```

```python
f = 2 * 7.2921e-5 * np.sin(np.deg2rad(curlz_tau.lat))
f = f.where(~(abs(curlz_tau.lat) < 1.5))
rho = 1000
f.plot()
```

```python
import cmocean
import cartopy
from matplotlib import pyplot as plt
```

```python
w_ek = (1/f/rho) * curlz_tau * 365 * 24 * 3600
```

```python
fig, ax = plt.subplots(
    3, 2,
    subplot_kw={"projection": cartopy.crs.PlateCarree()},
    figsize=(11, 7),
)

for t, _ax in zip(range(5), ax.flatten()):
    w_ek.coarsen(lon=3, lat=3, boundary="pad").mean().isel(time=4+t).plot(
        vmin=-3000, vmax=3000, levels=31,
        cmap=cmocean.cm.curl_r,
        ax=_ax, transform=cartopy.crs.PlateCarree(),
    )
    _ax.coastlines()
    _ax.set_extent([101, 108, 1, 7]);
```

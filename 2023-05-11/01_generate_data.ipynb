```python
import numpy as np
import xarray as xr
```

```python
# parameters

Nx = 360
Ny = 180 
Nt = 365

da = 1
a = 3
db = 2
b = -5

random_seed = 12345
```

```python
np.random.seed(random_seed)
```

```python
ds = xr.Dataset(
    {
        "A": (("t", "y", "x"), a + da * np.random.normal(size=(Nt, Ny, Nx))),
        "B": (("t", "y", "x"), b + db * np.random.normal(size=(Nt, Ny, Nx))),
    },
    coords={
        "t": ("t", np.arange(Nt)),
        "y": ("y", np.arange(Ny)),
        "x": ("x", np.arange(Nx)),
    }
)

ds
```

```python
ds.to_netcdf("data.nc")
```

```python
ds.A.plot.hist()
ds.B.plot.hist();
```

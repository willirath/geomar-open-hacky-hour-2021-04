```python
import xarray as xr
import numpy as np
```

```python
data = xr.DataArray(
    np.random.normal(size=(180, 360)),
    dims=("y", "x"),
    name="random_noise",
).to_dataset()
data
```

```python
data.random_noise.plot()
```

```python
data.random_noise.sel(x=slice(180, None), y=slice(90, None)).plot()
```

```python
data
```

```python
data = data.assign_coords(x=np.arange(data.dims["x"]))
data = data.assign_coords(y=np.arange(data.dims["y"]))
data
```

```python
data.random_noise.plot()
```

```python
data.random_noise.sel(x=slice(180, None), y=slice(90, None)).plot()
```

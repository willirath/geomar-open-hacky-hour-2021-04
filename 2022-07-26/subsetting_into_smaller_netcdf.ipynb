```python
import numpy as np
import xarray as xr
```

```python
dataset = xr.DataArray(
    np.random.normal(size=(365, 180, 360)),
    dims=("t", "lat", "lon"),
    name="data", 
).to_dataset()
dataset
```

```python
dataset.nbytes / 1e6
```

```python
dataset.sel(t=slice(0, 10), lat=slice(None, None, 10), lon=slice(None, None, 10))
```

```python
(
    dataset
    .sel(t=slice(0, 10), lat=slice(None, None, 10), lon=slice(None, None, 10))
    .to_netcdf("test.nc")
)
```

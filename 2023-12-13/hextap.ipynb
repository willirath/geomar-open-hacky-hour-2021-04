## Set up projection

(1000m hexes centred about Kiel.)

```python
hex_proj = hextraj.HexProj(
    lon_origin=10.14,
    lat_origin=54.32,
    hex_size_meters=1_000,
)

hex_proj
```

## Aux functions

- create a dict of hex corners for use in `hv.Polygons`
- aggregate a list of hexes and return a collection of rendered polygons
- aggregate a list of points nd return a collection of rendered points

```python
def hex_from_xy(x, y, data=0):
    """Create polyvon dict compatible with hv.Polygons."""
    _hex = hex_proj.lon_lat_to_hex_SoA(lon=x, lat=y)
    _corners = hex_proj.hex_corners_lon_lat(_hex)
    return {
        "x": [_c[0] for _c in _corners],
        "y": [_c[1] for _c in _corners],
        "data": data,
    }
```

```python
_hexes_aggregated = []
def stream_to_hex_centres(x, y):
    _hex = hex_proj.lon_lat_to_hex_SoA(lon=x, lat=y)
    _x, _y = hex_proj.hex_to_lon_lat_SoA(_hex)
    _hexes_aggregated.append(hex_from_xy(x, y, data=len(_hexes_aggregated)))
    return hv.Polygons(_hexes_aggregated, vdims="data")
```

```python
_pts_aggregated = []
def stream_to_pts(x, y):
    _pts_aggregated.append({"x": x, "y": y})
    return hv.Points(_pts_aggregated)
```

## Set up empty canvas and stream

- the `empty_canvas` is for capturing the clicks / taps
- the stream captures the taps in a list of x and y positions

```python
empty_canvas = hv.Points([])
stream = hv.streams.Tap(source=empty_canvas, x=hex_proj.lon_origin, y=hex_proj.lat_origin)
```

## Display empty canvas an dynamically map streams to point / hex vis

```python
(
    empty_canvas
    * hv.DynamicMap(stream_to_hex_centres, streams=[stream]).opts(cmap="viridis", clim=(0, 9))
    * hv.DynamicMap(stream_to_pts, streams=[stream])
)
```
